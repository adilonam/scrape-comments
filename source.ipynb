```python
import re

def extract_text_from_html(html):
    # This regex will match text within any HTML tag
    pattern = re.compile(r'>[^<]+<')
    matches = pattern.findall(html)
    # Remove the leading '>' and trailing '<' from each match
    text_list = [match[1:-1] for match in matches]
    return text_list

# Example usage
html = "<div>Hello</div><p>World</p><span>!</span>"
text_list = extract_text_from_html(html)
print(text_list)  # Output: ['Hello', 'World', '!']
```

```python
with open('./source.txt', 'r') as file:
    content = file.read()

text_list = extract_text_from_html(content)
print( " ".join(text_list) )  # Output: 'Hello World !'
```

```python
%reset -f
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# Read the HTML content from source.txt
with open('./source.txt', 'r') as file:
    html_content = file.read()

# Set up Selenium WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
print(html_content[:100])
# Load the HTML content into the WebDriver
driver.get("data:text/html;charset=utf-8," + html_content)

# Find elements by XPath
elements = driver.find_elements(By.XPATH, '//div[contains(@style, "text-align:")]')
print(elements)
# Print the text of each element
for element in elements:
    print(element.text)

# Close the WebDriver
driver.quit()
```
